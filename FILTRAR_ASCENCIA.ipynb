```python
import pandas as pd

# El nombre del archivo debe coincidir con el que se haya subido
file_name = 'ASCENSIA_DIABETES_COM_FIN_202501_DATOS Alicia.xlsx'

df = pd.read_excel(file_name)


# Ver las primeras filas del DataFrame
df.head()

```

```python
numerosaborrar = 'NumerosABorrar.xlsx'
aborrar = pd.read_excel(numerosaborrar, header= None).squeeze().tolist()

print(aborrar)
```

```python
# Filtrar el DataFrame para eliminar las filas cuyo 'ID_DATOS' esté en la lista de números
df_filtrado = df[~df['ID'].isin(aborrar)]

print(df_filtrado.head())  # Para verificar que las filas se han eliminado correctamente
```

```python
df['FECHA'] = pd.to_datetime(df['FECHA'], errors='coerce')
```

```python
fecha_inicio = pd.to_datetime('2020-04-01')
fecha_fin = pd.to_datetime('2025-03-19')
```

```python
# Filtrar las filas que están dentro del rango de fechas
df_filtrado = df_filtrado[(df_filtrado['FECHA'] >= fecha_inicio) & (df_filtrado['FECHA'] <= fecha_fin)]

# Paso 6: Guardar el DataFrame filtrado en un nuevo archivo Excel
df_filtrado.to_excel('datos_filtrados.xlsx', index=False)

print("El archivo filtrado se ha guardado exitosamente.")
```
