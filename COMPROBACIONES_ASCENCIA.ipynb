# **OBTENER DIFERENCIA ENTRE ID Y ID_DATOS, DIFERENCIA EN FECHA FISCAL ENTRE FECHA Y FECHA LLAMADA**

```python
# Convertir las columnas de fechas a solo fecha (sin hora)
archivo1['FECHA_LLAMADA'] = pd.to_datetime(archivo1['FECHA_LLAMADA']).dt.date
archivo2['FECHA'] = pd.to_datetime(archivo2['FECHA']).dt.date
```

```python
# Función para determinar el año fiscal basado en la fecha
def obtener_año_fiscal(fecha):
    # Si la fecha es antes del 1 de abril, el año fiscal es el año de la fecha
    # Si es después del 1 de abril, el año fiscal es el año siguiente
    if fecha.month >= 4:
        return fecha.year
    else:
        return fecha.year - 1
```

```python
# Aplicar la función para obtener el año fiscal para ambas fechas
archivo1['año_fiscal_llamada'] = archivo1['FECHA_LLAMADA'].apply(obtener_año_fiscal)
archivo2['año_fiscal'] = archivo2['FECHA'].apply(obtener_año_fiscal)
```

```python
# Verificar si ID en archivo1 no está en archivo2 y viceversa
id_no_en_archivo2 = archivo1[(archivo1['TOTAL_PEDIDO'].notna()) & (archivo1['TOTAL_PEDIDO'] != 0) & (archivo1['TOTAL_PEDIDO'] != '') & (~archivo1['ID'].isin(archivo2['ID_DATOS']))]
id_no_en_archivo1 = archivo2[~archivo2['ID_DATOS'].isin(archivo1['ID'])]
```

```python
# Comparar las fechas cuando los IDs están presentes en ambos archivos
# Realizamos una fusión entre los dos DataFrames para compararlos
merged_df = pd.merge(archivo1, archivo2, how='inner', left_on='ID', right_on='ID_DATOS', suffixes=('_archivo1', '_archivo2'))
```

```python
# Filtrar las filas donde los años fiscales no coinciden
diferencias_año_fiscal = merged_df[merged_df['año_fiscal_llamada'] != merged_df['año_fiscal']]
```

# **FARMACIAS GANADAS Y PERDIDAS**

```python
# Convertir la columna FECHA_LLAMADA a tipo datetime si no lo está
archivo1['FECHA_LLAMADA'] = pd.to_datetime(archivo1['FECHA_LLAMADA'])

# Crear una nueva columna 'Año_Fiscal' en base a la 'FECHA_LLAMADA'
def obtener_anno_fiscal(fecha):
    # Año fiscal de abril a marzo
    if fecha.month >= 4:
        return fecha.year  # Si es de abril a diciembre, es el año en curso
    else:
        return fecha.year - 1  # Si es de enero a marzo, es el año fiscal anterior

archivo1['Año_Fiscal'] = archivo1['FECHA_LLAMADA'].apply(obtener_anno_fiscal)

# Filtrar las farmacias con 'TOTAL_PEDIDO' distinto de 0 y no nulo
facturacion_df = archivo1.dropna(subset=['TOTAL_PEDIDO'])
facturacion_df = facturacion_df[facturacion_df['TOTAL_PEDIDO'] != 0]

# Obtener todos los años fiscales disponibles en el conjunto de datos
anos_fiscales = sorted(facturacion_df['Año_Fiscal'].unique())

# Crear un diccionario para almacenar los resultados por año fiscal
resultados = {}

# Iterar sobre los años fiscales para calcular las farmacias ganadas y perdidas por año
for anno_fiscal_actual in anos_fiscales:
    # Obtener el año fiscal anterior
    anno_fiscal_anterior = anno_fiscal_actual - 1

    # Filtrar las farmacias para el año fiscal actual y anterior
    farmacias_actuales = facturacion_df[facturacion_df['Año_Fiscal'] == anno_fiscal_actual]['CODIGO_CRM_FARMACIA'].unique()
    farmacias_anteriores = facturacion_df[facturacion_df['Año_Fiscal'] == anno_fiscal_anterior]['CODIGO_CRM_FARMACIA'].unique()

    # Encontrar las farmacias ganadas (están en el año fiscal actual pero no en el anterior)
    farmacias_ganadas = set(farmacias_actuales) - set(farmacias_anteriores)

    # Encontrar las farmacias perdidas (están en el año fiscal anterior pero no en el actual)
    farmacias_perdidas = set(farmacias_anteriores) - set(farmacias_actuales)

    # Almacenar los resultados en el diccionario
    resultados[anno_fiscal_actual] = {
        'Farmacias ganadas': len(farmacias_ganadas),
        'Farmacias perdidas': len(farmacias_perdidas)
    }

# Mostrar los resultados por año fiscal
for anno_fiscal, resultado in resultados.items():
    print(f"Año Fiscal {anno_fiscal}:")
    print(f"  Farmacias ganadas: {resultado['Farmacias ganadas']}")
    print(f"  Farmacias perdidas: {resultado['Farmacias perdidas']}")
    print("-" * 40)
```
